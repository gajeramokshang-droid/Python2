```python
# You are given a CSV file named ipl-matches.csv, which contains data of all Indian Premier League (IPL) matches played from the
# year 2008 to 2022.Read the dataset and display the basic information about the DataFrame.
# Using Python and Pandas, write code to answer the following questions:
# 1. How many matches were won by Gujarat Titans in a Super Over ?
# 2. How Many Matches won by Royal Challengers Bangalore Vs Kolkata Knight Riders.
# 3. How many Matches won by Toss Winner of the Match.
# 4. In how many matches was the result decided by the D/L method and the match was won by wickets?
# 5. In how many matches Toss won by Rajasthan Royals & Won the Match.
# 6. In how many matches V Kohli declared player of the match against Chennai Super Kings.
# 7. Which Team is the Toss Winner of the Final match Played in Dubai?
# 8. List out

import pandas as pd
df=pd.read_csv("ipl-matches.csv")
df
```

```python
# 1. How many matches were won by Gujarat Titans in a Super Over ?
x1=df[(df['WinningTeam']=='Gujarat Titans') & (df['SuperOver'])=='Y']
x2=x1.shape[0]
print(x2,"MATCHES WON BY GT IN SO")
```

```python
# 2. How Many Matches won by Royal Challengers Bangalore Vs Kolkata Knight Riders.
x3=df[(df["Team1"]=="Royal Challengers Bangalore") & (df["Team2"]=="Kolkata Knight Riders") & ((df["WinningTeam"]=="Royal Challengers Bangalore") | (df["WinningTeam"]=="Kolkata Knight Riders	"))]
x=x3.shape[0]
print(x,"Matches held between them won by either")
```

```python
# 3. How many Matches won by Toss Winner of the Match.
c1 = df[df["TossWinner"] == df["WinningTeam"]]
c1
```

```python
# 4. In how many matches was the result decided by the D/L method and the match was won by wickets?
v1=df[(df['method']=='D/L') & (df["WonBy"]=="Wickets")]
v2=v1.shape[0]
v2
```

```python
# 5. In how many matches Toss won by Rajasthan Royals & Won the Match.
p1=df[(df["TossWinner"]=="Rajasthan Royals") & (df["WinningTeam"]=="Rajasthan Royals")]
p2=p1.shape[0]
p2
```

```python
# 6. In how many matches V Kohli declared player of the match against Chennai Super Kings.
e1=df[((df["Team1"]=="Chennai Super Kings")|(df["Team2"]=="Chennai Super Kings")) & (df["Player_of_Match"]=="V Kohli")]
print(e1.shape[0],"POM by V Kohli")
```

```python
# 7. Which Team is the Toss Winner of the Final match Played in Dubai?
l1 = df[(df["MatchNumber"]=="Final") & (df["City"]=="Dubai")]
print(l1["TossWinner"].values)
```
