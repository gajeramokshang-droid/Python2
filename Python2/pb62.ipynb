```python
# Use the file ipl-matches.csv which contains data of all the IPL matches from year 2008 to 2022. Read this csv file and display the
# basic information like memory and data types for this data frame. Write python code for the following cases:
# 1. List out all matches gone in superover.
# 2. How Many Matches won by Chennai Super Kings at Kolkata.
# 3. In How Many Matches MS Dhoni is Player of Match Vs Mumbai Indians.
# 4. Display list of all matches in which Gujarat Titans won the Toss and Elected to Bat and won the match.
# 5. Display list of all matches won by Gujarat Titans.

import pandas as pd
df=pd.read_csv("ipl-matches.csv")
df
```

```python
# 1. List out all matches gone in superover.
x1=df["SuperOver"]
x1
```

```python
# 2. How Many Matches won by Chennai Super Kings at Kolkata.
matches = df[(df['City'] == 'Kolkata') & (df['WinningTeam'] == 'Chennai Super Kings')]
print(matches)
count = matches.shape[0]
print("Chennai Super Kings won", count, "matches at Kolkata")
```

```python
# 3. In How Many Matches MS Dhoni is Player of Match Vs Mumbai Indians.
x1 = df[(df["Team1"] == "Mumbai Indians") | (df["Team2"] == "Mumbai Indians") & df['Player_of_Match']==' MS Dhoni']
print(x1) 
```

```python
# 4. Display list of all matches in which Gujarat Titans won the Toss and Elected to Bat and won the match.
x2=df[((df["Team1"]=="Gujarat Titans") | (df["Team2"]=="Gujarat Titans")) & (df['TossWinner']=="Gujarat Titans") & (df["TossDecision"]=="bat") & (df["WinningTeam"]=="Gujarat Titans")]
x2
```

```python
# 5. Display list of all matches won by Gujarat Titans.
x3=df[df["WinningTeam"]=="Gujarat Titans"]
x3
```
