```python
# Write a Pandas program to calculate the multiply of length, width and depth for each cut of diamonds DataFrame.
df["Cal"] = df["x"] * df["y"] * df["z"]
df.groupby("cut")["Cal"].mean()
```

```python
# Write a Pandas program to merge two DataFrames on a single column.
color_info = pd.DataFrame({
    'color': ['E', 'I', 'J'],
    'rating': ['High', 'Medium', 'Low']
})

merged = df.merge(color_info, how='outer', indicator=True)
print(merged.head())
```
