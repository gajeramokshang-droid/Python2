```python
# 5. Do Height and Weight contain any outliers? (1 mark)
Q1=df["Height"].quantile(0.25)
Q3=df["Weight"].quantile(0.75)
Q1,Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_bound=Q1-1.5*IQR
higher_bound=Q3+1.5*IQR
lower_bound,higher_bound
```
